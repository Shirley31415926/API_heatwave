❗❗❗2023 Sentimental Analysis❗❗❗

```python
# Step 2: Import libraries
import json
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Step 3: Load and process the JSONL file
# Upload your file to Colab first
posts = []
with open('/content/drive/MyDrive/heat_posts2023.jsonl', 'r') as f:  ### Replace with your actual file path
    for line in f:
        posts.append(json.loads(line))

# Convert to DataFrame
df = pd.DataFrame(posts)

# Convert created_at to datetime
df['created_at'] = pd.to_datetime(df['created_at'])

# Step 4: Perform sentiment analysis
def get_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity  # Returns a float between -1 (negative) and 1 (positive)

df['sentiment'] = df['text'].apply(get_sentiment)

# Step 5: Analyze sentiment over time
# Set the datetime as index
df.set_index('created_at', inplace=True)

# Resample by time period (e.g., daily, weekly)
daily_sentiment = df['sentiment'].resample('D').mean()
weekly_sentiment = df['sentiment'].resample('W').mean()

# Step 6: Visualize the results
plt.figure(figsize=(15, 8))

# Daily sentiment
plt.subplot(2, 1, 1)
daily_sentiment.plot(title='Daily Average Sentiment', color='blue')
plt.ylabel('Sentiment Score')
plt.grid(True)

# Weekly sentiment
plt.subplot(2, 1, 2)
weekly_sentiment.plot(title='Weekly Average Sentiment', color='green')
plt.ylabel('Sentiment Score')
plt.grid(True)

plt.tight_layout()
plt.show()

# Additional visualization: Sentiment distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['sentiment'], bins=30, kde=True)
plt.title('Distribution of Sentiment Scores')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.show()

# Show some sample posts with their sentiment scores
print("\nSample Posts with Sentiment Scores:")
print(df[['text', 'sentiment']].sample(5))

# Save daily_sentiment as a CSV file
daily_sentiment.to_csv('/content/2023daily_sentiment.csv', index_label='Date', header=['Average_Sentiment'])
```
