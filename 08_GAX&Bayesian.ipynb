Due to the extremely low explanatory power of the regression models shown in File 06—with global average temperature and humidity exhibiting almost no ability to account for variations in sentiment scores (R² = 0.001)—the impact of UK temperature fluctuations on sentiment in both 2023 and 2024 was also found to be minimal. Even after applying a standardized 10-day moving average analysis, the explanatory power of the regression remained weak.


In File 08, we employed more advanced machine learning methods, with the following analyses:

***Generalized Additive Model (GAM)***

We constructed a GAM using the pyGAM library in Python, modeling the daily average sentiment score of social media posts as the dependent variable. Predictors included post count (as a proxy for public engagement), UK average daily temperature (Uktemp), a binary heatwave day indicator (heatwaveday), day of the year (day_of_year, to capture seasonal trends), and an interaction term between temperature and heatwave (Uktemp × heatwaveday). Continuous variables were modeled using penalized splines (s()), allowing for flexible nonlinear effects. The interaction term was manually constructed to avoid identifiability issues associated with tensor product terms.

***Bayesian GAM Model***

Given the relatively small sample size, we extended the model using a Bayesian framework implemented with the bambi and PyMC libraries. Predictors were modeled using B-spline basis functions via bs(), with degrees of freedom set to 4 (df=4). The model assumed a Gaussian likelihood and used default weakly informative priors. Posterior distributions were estimated using MCMC sampling (2 chains, 1000 draws per chain, 1000 warm-up iterations), and visualizations were produced with arviz.

The Bayesian model provided posterior estimates and credible intervals for each predictor, reducing reliance on p-values. This approach allowed for greater flexibility in handling nonlinear effects and interactions, offering a more comprehensive understanding of how climate conditions shape public sentiment expression online.

```python
from pygam import LinearGAM, s, f
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

df = pd.read_csv('/content/GAX2023.csv')
df['date'] = pd.to_datetime(df['date'])
df['day_of_year'] = df['date'].dt.dayofyear
df['temp_x_heat'] = df['Uktemp'] * df['heatwaveday']
df = df.dropna()

X = df[['post_count', 'Uktemp', 'heatwaveday', 'day_of_year', 'temp_x_heat']]
y = df['Average_Sentiment']

gam = LinearGAM(s(0) + s(1) + f(2) + s(3) + s(4)).fit(X, y)
gam.summary()

plt.figure(figsize=(18, 6))
labels = ['Post Count', 'Uktemp', 'Heatwave Day', 'Day of Year', 'Uktemp × Heatwave']

for i, label in enumerate(labels):
    plt.subplot(2, 3, i+1)
    XX = gam.generate_X_grid(term=i)
    plt.plot(XX[:, i], gam.partial_dependence(term=i, X=XX))
    plt.title(f'Effect of {label}')
    plt.xlabel(label)
    plt.ylabel('Partial Effect')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import bambi as bmb
import arviz as az
df = pd.read_csv('/content/GAX2023.csv')
df['date'] = pd.to_datetime(df['date'])
df['day_of_year'] = df['date'].dt.dayofyear
df['temp_x_heat'] = df['Uktemp'] * df['heatwaveday']
df = df.dropna()
model = bmb.Model(
    "Average_Sentiment ~ bs(post_count, df=4) + bs(Uktemp, df=4) + heatwaveday + bs(day_of_year, df=4) + bs(temp_x_heat, df=4)",
    data=df,
    family='gaussian'
)

results = model.fit(draws=1000, tune=1000, chains=2, random_seed=42)

az.summary(results)
az.plot_forest(results, kind="ridgeplot", combined=True)
```
