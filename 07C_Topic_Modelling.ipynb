```python
 # ---- 第一步：安装并加载必要的R包 ----
install.packages("stm")
install.packages("quanteda")
install.packages("textclean")
install.packages("jsonlite")
install.packages("readr")
install.packages("stringr")
install.packages("tm")   # 额外辅助处理

library(stm)
library(quanteda)
library(textclean)
library(jsonlite)
library(readr)
library(stringr)
library(tm)
```

## 2023

```python
# 1. 读数据
json_data <- stream_in(file("/content/heat_posts2023.jsonl"))
texts     <- json_data$text

# 2. 文本清洗
texts_clean <- texts %>%
  replace_html() %>%
  replace_tag() %>%
  replace_symbol() %>%
  replace_emoji() %>%
  replace_contraction() %>%
  str_remove_all("[[:punct:]]") %>%
  str_remove_all("[[:digit:]]") %>%
  tolower() %>%
  str_squish()

# 3. Tokens
toks <- tokens(texts_clean,
               remove_punct   = TRUE,
               remove_numbers = TRUE,
               remove_symbols = TRUE) %>%
  tokens_tolower() %>%
  tokens_remove(stopwords("en")) %>%
  tokens_wordstem("en") %>%
  tokens_keep(min_nchar = 3)

# 4. DFM & 放宽 dfm_trim
dtm <- dfm(toks) %>%
  dfm_trim(min_docfreq = 1, min_termfreq = 2) %>%  # <-- 放宽到几乎不过滤
  dfm_select(pattern = "^[a-z]{3,}$", valuetype = "regex")

# 5. 检查非空文档数
message("Non-empty docs: ", sum(rowSums(dtm) > 0))

# 6. STM
out <- convert(dtm, to = "stm")
set.seed(12345)
model <- stm(
  documents = out$documents,
  vocab     = out$vocab,
  K         = 6,
  init.type = "Spectral",
  LDAbeta   = FALSE,
  verbose   = TRUE
)

# 7. 主题词 & 思考片段
labelTopics(model, n = 12)
keep_idx   <- out$docs$doc_id
keep_texts <- texts_clean[keep_idx]
print(findThoughts(model, texts = keep_texts, n = 3))

# 8. 可视化
plot(model, type = "summary")
plot(model, type = "labels", labeltype = "frex")
```
